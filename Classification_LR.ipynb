#split in x and y       

v58 is target(Y)

```python
X = df.drop('V58',axis =1)
Y = df['V58']
```

```python
X.shape,Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=16,stratify=Y)
```

stratifed train and split method for random sampling

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
from sklearn.naive_bayes import BernoulliNB, GaussianNB, MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

```

```python
lr = LogisticRegression(class_weight='balanced',random_state=16,solver = 'saga')
```

```python
lr.fit(X_train,Y_train)
```

the fit transform calculated the coefficents used for calculating the linaer eqaution required for linear regression below the sigmoid function
more or higher value of the coeffients more is the weightage or importance of that independent variable

```python
lr.coef_
```

```python
lr.coef_.shape
```

```python
Y_pred_lr = lr.predict(X_test)
```

```python
Y_test.shape,Y_pred_lr.shape
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_pred_lr))
```

```python
svm_model = SVC(class_weight='balanced',random_state=16)
```

```python
svm_model.fit(X_train,Y_train)
```

```python
svm_model.coef_
```

#Naive Bayes

```python
nb_ber = BernoulliNB()
nb_g = GaussianNB()
nb_m = MultinomialNB()
```

```python
nb_ber.fit(X_train,Y_train)
nb_g.fit(X_train,Y_train)
nb_m.fit(X_train,Y_train)
```

```python
Y_predict_ber = nb_ber.predict(X_test)
Y_predict_g = nb_g.predict(X_test)
Y_predict_m = nb_m.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print("****************BernoulliNB**********************")
print(classification_report(Y_test,Y_predict_ber))
print("****************GaussianNB***********************")
print(classification_report(Y_test,Y_predict_g))
print("****************MultinomialNB*********************")
print(classification_report(Y_test,Y_predict_m))
```
